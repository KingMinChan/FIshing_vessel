```python
from PIL import Image
import os

# 이미지가 있는 폴더 경로 설정
image_folder = '테스트셋'
output_folder = 'crop_test_image'

# 저장할 폴더가 없으면 생성
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 이미지가 저장된 폴더에서 파일 읽기
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png', '.jpeg'))]

# 이미지 처리
for image_file in image_files:
    # 이미지 파일 경로
    image_path = os.path.join(image_folder, image_file)
    
    # Pillow로 이미지 열기
    image = Image.open(image_path)
    
    # 바운딩 박스 좌표 (예시)
    # 실제 좌표는 YOLO 또는 SSD 등의 탐지 결과에서 얻어야 함
    width, height = image.size
    xmin, ymin, xmax, ymax = int(width * 0.1), int(height * 0.1), int(width * 0.8), int(height * 0.8)
    
    # 이미지 자르기 (left, upper, right, lower)
    cropped_image = image.crop((xmin, ymin, xmax, ymax))
    
    # 자른 이미지를 저장할 경로 설정 (원본 파일명 유지)
    cropped_image_path = os.path.join(output_folder, image_file)
    
    # 자른 이미지 저장
    cropped_image.save(cropped_image_path)
    
    print(f"{image_file}에서 객체를 잘라 저장했습니다: {cropped_image_path}")

print("모든 이미지 처리 완료.")
```
