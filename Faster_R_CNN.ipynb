```python
import os
import random
import shutil

def train_test_split(src_folder, train_size=0.8):
    # 기존 폴더를 삭제하고 깨끗한 상태로 시작한다.
    shutil.rmtree(src_folder+'Train/범장망', ignore_errors=True)
    shutil.rmtree(src_folder+'Test/범장망', ignore_errors=True)
    shutil.rmtree(src_folder+'Train/타망', ignore_errors=True)
    shutil.rmtree(src_folder+'Test/타망', ignore_errors=True)
    shutil.rmtree(src_folder+'Train/유망', ignore_errors=True)
    shutil.rmtree(src_folder+'Test/유망', ignore_errors=True)
    shutil.rmtree(src_folder+'Train/안강망', ignore_errors=True)
    shutil.rmtree(src_folder+'Test/안강망', ignore_errors=True)
    shutil.rmtree(src_folder+'Train/통발', ignore_errors=True)
    shutil.rmtree(src_folder+'Test/통발', ignore_errors=True)

    # train 폴더와 test 폴더를 새로 생성한다.
    os.makedirs(src_folder+'Train/범장망')
    os.makedirs(src_folder+'Test/범장망')
    os.makedirs(src_folder+'Train/타망')
    os.makedirs(src_folder+'Test/타망')
    os.makedirs(src_folder+'Train/유망')
    os.makedirs(src_folder+'Test/유망')
    os.makedirs(src_folder+'Train/안강망')
    os.makedirs(src_folder+'Test/안강망')
    os.makedirs(src_folder+'Train/통발')
    os.makedirs(src_folder+'Test/통발')

    _, _, 범장망_images = next(os.walk(src_folder + '/범장망'))
    
    num_범장망_images = len(범장망_images)
    num_범장망_images_train = int(train_size * num_범장망_images)
    num_범장망_images_test = num_범장망_images - num_범장망_images_train

    # 이미지를 무작위로 골라 train 폴더와 test 폴더로 나눈다.
    범장망_train_images = random.sample(범장망_images, num_범장망_images_train)
    for img in 범장망_train_images:
        shutil.copy(src=src_folder+'범장망/'+img, dst=src_folder+'Train/범장망/')
    범장망_test_images = [img for img in 범장망_images if img not in 범장망_train_images]
    for img in 범장망_test_images:
        shutil.copy(src=src_folder+'범장망/'+img, dst=src_folder+'Test/범장망/')

    _, _, 안강망_images = next(os.walk(src_folder + '/안강망'))
    
    num_안강망_images = len(안강망_images)
    num_안강망_images_train = int(train_size * num_안강망_images)
    num_안강망_images_test = num_안강망_images - num_안강망_images_train

    # 이미지를 무작위로 골라 train 폴더와 test 폴더로 나눈다.
    안강망_train_images = random.sample(안강망_images, num_안강망_images_train)
    for img in 안강망_train_images:
        shutil.copy(src=src_folder+'안강망/'+img, dst=src_folder+'Train/안강망/')
    안강망_test_images = [img for img in 안강망_images if img not in 안강망_train_images]
    for img in 안강망_test_images:
        shutil.copy(src=src_folder+'안강망/'+img, dst=src_folder+'Test/안강망/')
    
    _, _, 유망_images = next(os.walk(src_folder + '/유망'))
    
    num_유망_images = len(유망_images)
    num_유망_images_train = int(train_size * num_유망_images)
    num_유망_images_test = num_유망_images - num_유망_images_train

    # 이미지를 무작위로 골라 train 폴더와 test 폴더로 나눈다.
    유망_train_images = random.sample(유망_images, num_유망_images_train)
    for img in 유망_train_images:
        shutil.copy(src=src_folder+'유망/'+img, dst=src_folder+'Train/유망/')
    유망_test_images = [img for img in 유망_images if img not in 유망_train_images]
    for img in 유망_test_images:
        shutil.copy(src=src_folder+'유망/'+img, dst=src_folder+'Test/유망/')

    _, _, 타망_images = next(os.walk(src_folder + '/타망'))
    
    num_타망_images = len(타망_images)
    num_타망_images_train = int(train_size * num_타망_images)
    num_타망_images_test = num_타망_images - num_타망_images_train

    # 이미지를 무작위로 골라 train 폴더와 test 폴더로 나눈다.
    타망_train_images = random.sample(타망_images, num_타망_images_train)
    for img in 타망_train_images:
        shutil.copy(src=src_folder+'타망/'+img, dst=src_folder+'Train/타망/')
    타망_test_images = [img for img in 타망_images if img not in 타망_train_images]
    for img in 타망_test_images:
        shutil.copy(src=src_folder+'타망/'+img, dst=src_folder+'Test/타망/')

    _, _, 통발_images = next(os.walk(src_folder + '/통발'))
    
    num_통발_images = len(통발_images)
    num_통발_images_train = int(train_size * num_통발_images)
    num_통발_images_test = num_통발_images - num_통발_images_train

    # 이미지를 무작위로 골라 train 폴더와 test 폴더로 나눈다.
    통발_train_images = random.sample(통발_images, num_통발_images_train)
    for img in 통발_train_images:
        shutil.copy(src=src_folder+'통발/'+img, dst=src_folder+'Train/통발/')
    통발_test_images = [img for img in 통발_images if img not in 통발_train_images]
    for img in 통발_test_images:
        shutil.copy(src=src_folder+'통발/'+img, dst=src_folder+'Test/통발/')
```
